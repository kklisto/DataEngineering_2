```python
import feedparser
import pandas as pd

# Récupération du flux RSS
rss_url = 'https://www.lemonde.fr/rss/une.xml'
feed = feedparser.parse(rss_url)

# Conversion du flux RSS en DataFrame
rss_df = pd.DataFrame(feed.entries)
```

```python
rss_df = rss_df.iloc[:10,]
rss_df
```

```python
# Export de la DataFrame en CSV
rss_df.to_csv('lemonde_rss.csv', index=False)
```
